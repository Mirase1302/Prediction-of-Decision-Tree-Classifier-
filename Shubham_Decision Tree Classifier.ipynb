```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
sub_type=pd.read_csv("iris.csv")
sub_type
```

```python
in_file=sub_type.drop(columns=["Id","Species"])
in_file
```

```python
out_file=sub_type["Species"]
out_file
```

```python
AI_Model=DecisionTreeClassifier()
AI_Model.fit(in_file.values,out_file.values)
Prediction=AI_Model.predict([[2.9,4.5,7.3,9.2],[4.1,9.3,1.6,6.4],[5.9,3.6,1.1,0.5]])
Prediction
```
